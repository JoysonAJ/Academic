# Simulate Perceptron network for AND, OR and XOR gates.

```python
import numpy as np

def unitStep(v):
    if v>=0:
        return 1
    else:
        return 0

def perceptronModel(x,weight,bais):
    v = np.dot(x,weight)+bais
    return unitStep(v)

def AND_GATE(x):
    weight = np.array([1,1])
    bais = -1.5
    return perceptronModel(x,weight,bais)
    
def OR_GATE(x):
    weight = np.array([1,1])
    bais = -0.5
    return perceptronModel(x,weight,bais)
    
def NOT_GATE(x):
    weightNOT = -1
    baisNOT = 0.5
    return perceptronModel(x,weightNOT,baisNOT)
    
def XOR_GATE(x):
    y1 = AND_GATE(x)
    y2 = OR_GATE(x)
    y3 = NOT_GATE(y1)
    
    final_x = np.array([y1,y2])
    finalOutput = AND_GATE(final_x)
    return finalOutput

test1 =np.array([0,1]) 
test2 =np.array([1,1]) 
test3 =np.array([0,0]) 
test4 =np.array([1,0]) 

print("AND GATE\n")
print("___________________________________________")
print(AND_GATE(test1))
print(AND_GATE(test2))
print(AND_GATE(test3))
print(AND_GATE(test4))

print("\nOR GATE\n")
print("___________________________________________")
print(OR_GATE(test1))
print(OR_GATE(test2))
print(OR_GATE(test3))
print(OR_GATE(test4))

print("\nXOR GATE\n")
print("___________________________________________")
print(XOR_GATE(test1))
print(XOR_GATE(test2))
print(XOR_GATE(test3))
print(XOR_GATE(test4))
```

# Simulate Adaline network for bipolar OR gate.

```python
from numpy import array
features = array([
    [1,1],
    [1,-1],
    [-1,1],
    [-1,-1]
])
lables = array([-1,1,-1,-1])

# set the weight and bais random value not zeo
weight =[0.2,0.2]
bais = 0.2

# Alpha value
learning_rate = 0.2

epoch = 2

for i in range(epoch):
    
    finalError = 0.0
        
    for j in range(features.shape[0]):
        actual = lables[j]
        x1 = features[j][0]
        x2 = features[j][1]
        
        y = bais + (x1* weight[0] + x2*weight[1])
        print("The value is y",y)
        
        errorValue = actual - y
        
#         print(errorValue)
        weight[0] = weight[0] + learning_rate * (errorValue) * x1
        weight[1] = weight[1] + learning_rate * (errorValue) * x2
        
        print(f"The Updated wieght is {weight}")
        
        bais = bais + learning_rate* errorValue
        print(f"The updated bais is {bais}")
        
        finalError = errorValue * errorValue
        print(f"Error Result of {j+1} is {finalError}")
        
        print("________________________________________")
```
