```python
import cv2
import numpy as np
```

```python
# import os
# os.getcwd()
dirName = r'E:\Submission\WorkBranch\Academic\Sem2\Image Processing\Jupyter File\Images'
```

```python
image1 = cv2.imread(dirName+'\Image1.jpg')
image2 = cv2.imread(dirName+'\Image2.jpg')
```

```python
from matplotlib import pyplot as plt 
# %pylab inline
```

### ADD


```python
addImage = np.add(image1,image2)
cv2.imshow("Additon of 2 Images",addImage)
cv2.waitKey(0)
```

```python
# Open CV CODE

```

### Subtraction

```python
subImage = np.subtract(image1,image2)
cv2.imshow("Subtraction of 2 Images",subImage)
cv2.waitKey(0)
```

```python
# Refferance
plt.imshow(subImage)
plt.show()
```

### Logical Operator

```python
bitOr = cv2.bitwise_or(image1,image2,mask=None)
cv2.imshow("BitWise Or",bitOr)
cv2.waitKey(0)
```

```python
bitAnd = cv2.bitwise_and(image1,image2,mask=None)
cv2.imshow("BitWise And",bitAnd)
cv2.waitKey(0)
```

```python
bitXor = cv2.bitwise_xor(image1,image2,mask=None)
cv2.imshow("BitWise And",bitXor)
cv2.waitKey(0)
```

```python
bitNot_Image1 = cv2.bitwise_not(image1)
cv2.imshow("BitWise And",bitNot_Image1)
cv2.waitKey(0)
```

```python
bitNot_Image2 = cv2.bitwise_not(image2)
cv2.imshow("BitWise And",bitNot_Image2)
cv2.waitKey(0)
```
